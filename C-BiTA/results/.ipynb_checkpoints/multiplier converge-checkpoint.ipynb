```python
import matplotlib.pyplot as plt
import math
from matplotlib import font_manager
import pandas as pd
import seaborn as sns
```

```python
# def read_ite_gene(filename):
#     df1 = pd.DataFrame([["TNM_TAP", 0, 100000000, 1.0, 1, 0]], columns=["ObjID","Iter", "OFV", "StepSize", "ConvIndc", "Time"] )
#     efficient_line = []
#     with open(filename, 'r') as f1:
#         l1 = f1.readlines()
#         for i in range(len(l1)):
#             if l1[i][0]=='-':
#                 efficient_line.append(i)
#         l1 = l1[efficient_line[0]+1 : efficient_line[1]]
#     df = pd.read_table(filename,sep='\\s+', skiprows=efficient_line[0]+1, nrows=efficient_line[1]-efficient_line[0]-2)
#     df1 = df1.append(df, ignore_index=True)
#     return (df1)
def read_ite_gene(filename):
    df1 = pd.read_table(filename,sep='\s+')
    return (df1)
```

```python
fig = plt.figure(figsize=(8,7))
ax1 = fig.add_subplot(111)
ax1.set_yscale('log')
#palette=sns.color_palette("hls", len([0, 2, 6, 12, 20, 30, 45, 70, 100, 160]))
colors = ['mediumblue', 'dodgerblue',  'forestgreen', 'limegreen', 'gold', 'orange','red', 'brown']
linestyles=[':',  '-.', '--', '--', '-.', '-', '-', '-']
linewidths = [ 3, 1.5, 1.5, 3, 2.5, 1.5, 2.5, 3.5]
for k,i in enumerate([2, 6, 12, 20, 30, 60, 100, 160]):
    df = read_ite_gene("%s/chicagosketch_SBA.ite"%i)
    ax1.plot(df["Time"][0:], df["ConvIndc"][0:],label=r'$\alpha_{0}$=%s'%i, linestyle=linestyles[k], linewidth = linewidths[k], color=colors[k])
font1 = {'family':'Times New Roman','weight':'heavy','size': 20}
ax1.set_xlabel('CPU Time (s)',fontdict=font1)
ax1.set_ylabel('Relative Gap', fontdict=font1)
long = 500
ax1.set_xlim(-0.02*long, long)
ax1.set_ylim(pow(10, -11), 2)
ax1.set_yticks([ pow(10, -10), pow(10, -8), pow(10, -6), pow(10, -4), pow(10, -2), pow(10, 0)])
ticks_font = font_manager.FontProperties(family='STIXGeneral', style='normal',
    size=16, weight='heavy', stretch='normal')
for label in ax1.get_xticklabels():
    label.set_fontproperties(ticks_font)
for label in ax1.get_yticklabels():
    label.set_fontproperties(ticks_font)
ax1.legend(prop={'family' : 'STIXGeneral', 'size' : 16,'weight':'heavy'}, ncol=2)
ax1.grid(color='darkgrey',linestyle=':',linewidth=1)
plt.tight_layout()
fig.savefig('vot_compareBAA.eps',dpi=600,bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(4,2))
ax1 = fig.add_subplot(111)
ax1.set_yscale('log')
colors = ['mediumblue', 'dodgerblue',  'forestgreen', 'limegreen', 'gold', 'orange','red', 'brown']
linestyles=[':',  '-.', '--', '--', '-.', '-', '-', '-']
linewidths = [ 3, 1.5, 1.5, 3, 2.5, 1.5, 2.5, 3.5]
for k,i in enumerate([2, 6, 12, 20, 30, 60, 100, 160]):
    df = read_ite_gene("%s/chicagosketch_SBA.ite"%i)
    if ((i>=6) and (i<=30)):
        line = '--'
        linewidth = 2
    elif (i>30):
        line = '-'
        linewidth= 1.5
    else:
        line=':'
        linewidth = 3
    ax1.plot(df["Time"][0:], df["ConvIndc"][0:],label=r'$\alpha_{avg}$=%s'%i, linestyle=linestyles[k], linewidth = linewidths[k], color=colors[k])
font1 = {'family':'Times New Roman','weight':'heavy','size': 20}
ax1.set_xlim(-10, 130)
ax1.set_ylim(pow(10, -6), pow(10,-4))
ax1.set_yticks([pow(10, -6), pow(10, -4)])
ax1.set_xticks([])
ticks_font = font_manager.FontProperties(family='STIXGeneral', style='normal',
    size=16, weight='heavy', stretch='normal')
for label in ax1.get_xticklabels():
    label.set_fontproperties(ticks_font)
for label in ax1.get_yticklabels():
    label.set_fontproperties(ticks_font)
ax1.grid(color='darkgrey',linestyle=':',linewidth=1)
plt.tight_layout()
fig.savefig('miniBAA.eps',dpi=600,bbox_inches='tight')
plt.show()
```
